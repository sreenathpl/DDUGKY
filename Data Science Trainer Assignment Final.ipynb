## Merging 2 tables

```python
data_merge = ip_user_ID.merge(ip_rigorbuilder[['total_statements','total_reasons','uid']], how='left', left_on='User ID', right_on='uid').drop(columns=['uid'])
```

```python
# Check the merged table
data_merge.head()
```

```python
data_merge['Team Name'].unique()
```

We can see that the Team **BrandTech Lab** has also writen as **Brandtech Lab** which will consider as 2 team while grouping. So we have to update the value of one.

```python
data_merge['Team Name'][data_merge['User ID'] == 336] = 'BrandTech Lab'
```

```python
# Check the updated values.
data_merge['Team Name'].unique()
```

## Checking the data types of the fields

```python
data_merge.info()
```

We can see that the data types of **total_statements and total_reasons** are object data types. We have to convert them to numeric one.

```python
data_merge[['total_statements', 'total_reasons']] = data_merge[['total_statements', 'total_reasons']].astype(int)
```

```python
## Check the updated table.
data_merge.info()
```

## Leaderboard TeamWise

```python
lb_teamwise = data_merge.groupby(['Team Name']).mean().round(2).sort_values(by=['total_statements', 'total_reasons'], ascending=False)
lb_teamwise.reset_index(inplace=True)
lb_teamwise.columns = ['Thinking Teams Leaderboard','Average Statements','Average Reasons']
lb_teamwise.insert(loc = 0,
          column = 'Team Rank',
          value = [i+1 for i in range(len(lb_teamwise))])
```

```python
blankIndex=[''] * len(lb_teamwise)
lb_teamwise.index = blankIndex
```

**Dashboard of Leaderboard TeamWise (Output)**

```python
lb_teamwise
```

We can see adifference in the value of Average Reasons of "Tech Lab". Since the rank is decided on the basis of sum of average statements and reasons the answer is 10.

## Leaderboard Individual

```python
# Store the "Name & User ID" to another DF
name = data_merge[['Name','User ID']] 
```

```python
# Converting the "Name" to lower case for sorting easiness
data_merge['Name'] = data_merge['Name'].str.lower()
```

```python
lb_individual = data_merge[['S No','Name','User ID','total_statements','total_reasons']].sort_values(by=['total_statements', 'total_reasons','Name'],ascending=[False,False,True])
```

```python
lb_individual
```

```python
# For the below table is sorted like "total_reasons" in decending order and "Name" in ascending order.
lb_individual[14:]
```

```python
lb_individual[14:] = lb_individual[14:].sort_values(by=['total_reasons','Name'],ascending=[False,True])
```

```python
# Updating the 'S No'
lb_individual['S No'] = [i+1 for i in range(len(lb_individual))]
```

```python
lb_individual.reset_index(inplace = True, drop = True)
```

```python
lb_individual
```

We can see the values in the column "Name" is in lowercase. We have tochange it to orginal names.

```python
for i in range(len(lb_individual)):
    for j in range(len(name)):
        if name['User ID'][j] == lb_individual['User ID'][i]:
            lb_individual['Name'][i] = name['Name'][j]
```

```python
# Updating columns
lb_individual.columns = ['Rank','Name','UID','No. of Statements','No. of Reasons']
```
